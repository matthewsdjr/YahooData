```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame()
tickers = ["AAPL", "AMZN", "TSLA"]
for ticker in tickers:
    df[ticker] = yf.Ticker(ticker).history(start = "2020-01-01", end = "2021-01-01").Close

df.plot()
plt.show()
```
